## Eliminación de columnas / creación columnas / cambio nombre

```python
classes= classes.drop(classes.columns[0], axis=1)
classes = classes.drop('index', axis=1)
classes = classes.drop('_id', axis=1)
classes=classes.rename(columns={'hit_die': 'health'})
classes = classes.drop('proficiencies_', axis=1)
classes = classes.drop('proficiencies', axis=1)
classes=classes.rename(columns={'proficiencies2': 'proficiencies'})
classes = classes.drop('proficiency_choices', axis=1)
classes = classes.drop('saving_throws', axis=1)
classes = classes.drop('saving', axis=1)
classes=classes.rename(columns={'savingthrows': 'saving_throws'})
classes = classes.drop('starting_equipment', axis=1)
classes = classes.drop('start1', axis=1)
classes = classes.drop('classl', axis=1)
classes = classes.drop('class_levels', axis=1)
classes = classes.drop('url', axis=1)
classes = classes.drop('class1', axis=1)
classes = classes.drop('subclasses', axis=1)
classes=classes.rename(columns={'subc': 'subclasses'})
classes = classes.drop('spellcasting', axis=1)
classes=classes.rename(columns={'spellcasting_': 'spellcasting'})
classes=classes.rename(columns={'name': 'class_name'})
```

Esto más adelante lo optimicé para evitar crear columnas nuevas para luego borrarlas o renombrarlas.
