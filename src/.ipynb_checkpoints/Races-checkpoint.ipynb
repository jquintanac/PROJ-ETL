```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import warnings
warnings.filterwarnings('ignore')

import json
from pandas.io.json import json_normalize
```

```python
races = pd.read_csv('../data/races.csv')
```

# RACES CLEANING

```python
races.head()
```

## Languages

```python
races.languages
```

Transformo una string con formato lista de diccionarios con json y reemplazando las comillas.

```python
a = [json.loads(i.replace("'", '"')) for i in races.languages]
```

Separo los lenguajes teniendo en cuenta si hablan uno o dos

```python
lang=[]
for i in a:
    if len(i) ==2:
        for j in i:
            lang.append(j['name'])
    if len(i) ==1:
        for j in i:
            lang.append(j['name'])
            lang.append('No second language')
```

```python
first=[]
second=[]
for i in range(0, len(lang)):
    if i % 2:
        second.append(lang[i])
    else :
        first.append(lang[i])
```

```python
races['first_lang']=first
races['second_lang']=second
races = races.drop('languages', axis=1)
```

```python
races.language_options.fillna('None', inplace=True)
races.language_options[4]='Choose one more language'
races.language_options[7]='Choose one more language'
```

## Alignment

Se extrae el alineamiento de cada raza según la descripción de alineamiento en una tabla aparte que se vincula a traves de id_races. 

```python
races['id_races']=[i for i in range(1,len(races)+1)]
```

```python
lst=['law', 'chao', 'good', 'neutral', 'evil']
alignment=pd.DataFrame()

for e in lst:
    alignment[e]=races.alignment.apply(lambda x: e in x)

alignment['id_races']=races.id_races
```

```python
alignment=alignment.reindex(columns=['id_races','law','chao','neutral','good','evil'])

alignment
```

```python
alignment=alignment.rename(columns={'law': 'lawful'})
alignment=alignment.rename(columns={'chao': 'chaotic'})
```

```python
alignment['lawful'] = alignment['lawful'].map({True: 'True', False: 'False'})
alignment['chaotic'] = alignment['chaotic'].map({True: 'True', False: 'False'})
alignment['neutral'] = alignment['neutral'].map({True: 'True', False: 'False'})
alignment['good'] = alignment['good'].map({True: 'True', False: 'False'})
alignment['evil'] = alignment['evil'].map({True: 'True', False: 'False'})

alignment
```

Los humanos que no son legales/caóticos ni buenos/malos se consideran neutrales, así que se cambia el valor: primero en string y luego replace, para no trabajar con booleanos.

```python
alignment['neutral'][7]= alignment['neutral'][7].replace('False','True')


alignment
```

Añadimos el id de alignment a races para poder trazarla posteriormente.

```python
races['id_alignment']=[i for i in range(1,len(races)+1)]
```

## Bonus

```python
a = [json.loads(i.replace("'", '"')) for i in races.ability_bonuses]
```

```python
bonus=[]
bonus2=[]
for i in a:
    for j in i:
        bonus.append(j['name']+"+"+str(j['bonus']))
        if len(bonus)==len(i):
            bonus2.append(bonus)
            bonus=[]

bonus2
```

```python
races['bonus2']=bonus2
```

```python
bon=[]
for i in races.bonus2:
    i=str(i)
    bon.append(i)
bonu=[]
for i in bon:
    i=i.replace("[","").replace("]","").replace("]","").replace("'","")
    bonu.append(i)

races['bonus']=bonu
```

```python
races.ability_bonus_options[4]='Choose 2 from bonus'
races=races.rename(columns={'ability_bonus_options': 'bonus_option'})
races.bonus_option.fillna('None', inplace=True)
```

## Proficiencies

```python
a = [json.loads(i.replace("'", '"')) for i in races.starting_proficiencies]
```

```python
prof=[]
prof2=[]

for i in a:
    if len(i)== 0:
            prof2.append('None')
    else:
        for j in i:
            prof.append(j['name'])
            if len(prof)==len(i):
                prof2.append(prof)
                prof=[]

races['prof']=prof2

prof3=[]

for i in races.prof:
    i=str(i)
    prof3.append(i)
    
prof4=[]

for i in prof3:
    i=i.replace("[","").replace("]","").replace("]","").replace("'","")
    prof4.append(i)


races['proficiencies_start']=prof4
```

```python
races=races.drop(columns='prof', axis=1)
races=races.drop(columns='starting_proficiencies', axis=1)
```

## Traits

```python
races.traits
```

```python
a = [json.loads(i.replace("'", '"')) for i in races.traits]

trait=[]
traits=[]

for i in a:
    if len(i)== 0:
            traits.append('None')
    else:
        for j in i:
            trait.append(j['name'])
            if len(trait)==len(i):
                traits.append(trait)
                trait=[]

races['traits2']=traits

trait=[]

for i in races.traits2:
    i=str(i)
    trait.append(i)
    
trait2=[]

for i in trait:
    i=i.replace("[","").replace("]","").replace("]","").replace("'","")
    trait2.append(i)


races['traits_']=trait2
```

## Subraces

Extraigo subraces como columna nueva.

```python
a = [json.loads(i.replace("'", '"')) for i in races.subraces]

subr=[]

for i in a:
    if len(i)== 0:
            subr.append('None')
    else:
        for j in i:
            subr.append(j['name'])

races['subraces2']=subr
```

## Eliminación de columnas / creación columnas / cambio nombre 

```python
races = races.drop('index', axis=1)
races = races.drop('_id', axis=1)
races= races.drop(races.columns[0], axis=1)
races=races.drop(columns='traits', axis=1)
races=races.drop(columns='traits2', axis=1)
races=races.drop(columns='trait_options', axis=1)
races=races.drop(columns='starting_proficiency_options', axis=1)
races=races.rename(columns={'traits_': 'traits'})
races=races.drop(columns='subraces', axis=1)
races=races.rename(columns={'subraces2': 'subrace_name'})
races=races.drop(columns='alignment', axis=1)
races=races.drop(columns='language_desc', axis=1)
races=races.drop(columns='size_description', axis=1)
races=races.drop(columns='age', axis=1)
races=races.drop(columns='url', axis=1)
races=races.drop(columns='bonus2', axis=1)
races=races.drop(columns='ability_bonuses', axis=1)
races=races.rename(columns={'name': 'race_name'})
```

```python
races= races[['id_races', 'race_name', 'size', 'speed', 'id_alignment', 'first_lang', 'second_lang', 'language_options', 'bonus', 'bonus_option', 'subrace_name', 'proficiencies_start', 'traits']]

races
```
